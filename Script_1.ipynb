```python
import numpy as np
import numpy.ma as ma
import pandas as pd
import geopy.distance
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

#Get library
```

```python
farmer_data_2 = pd.read_csv("/content/Data/farmer_list_2.csv")
supplier_data_2 = pd.read_csv("/content/Data/supplier_list_2.csv")
# label_data = pd.read_csv("/content/Data/label.csv")
# label_data_list= np.array(label_data.drop(['Name'], axis=1).values.tolist())

input = [109,-72, 100,0,0,200,400,0]
#Get CSV
```

```python
commodity_score = np.zeros((100,100, 6))
for farmer_index, farmer_row in farmer_data_2.iterrows():
    for supplier_index, supplier_row in supplier_data_2.iterrows():
        for commodity_col in range(1, 7): 
            supplier_amount = supplier_row[f'Type Needed {commodity_col}']
            farmer_amount = farmer_row[f'Commodity {commodity_col}'] 
            if farmer_amount >= supplier_amount:
                score = 1 - (farmer_amount - supplier_amount)/500
                commodity_score[farmer_index][supplier_index][commodity_col - 1] = score
            else:
                commodity_score[farmer_index][supplier_index][commodity_col - 1] = 0
final_commodity_score = np.mean(commodity_score,axis=-1)
#Hitung commodity score
```

```python
print(final_commodity_score[0])
#Commodity score dari supplier 1 ke semua farmer
```

```python
lst_lat_long_train = farmer_data_2[['Lat','Long']] 
K_Means = KMeans(n_clusters=5, init='k-means++')
K_Means.fit(lst_lat_long_train)
farmer_data_2['Cluster'] = K_Means.predict(farmer_data_2[['Lat', 'Long']])
#Predict using K Mean
labeled_farmer_data = farmer_data_2.copy()
print(K_Means.cluster_centers_)
```

```python
#Prediction Based on how far from cluster center

def findNearestCluster(coor_supp):
  centroids = K_Means.cluster_centers_
  N, M = centroids.shape
  lst_coor = np.zeros(N)
  for i, data in enumerate(centroids):
    lst_coor[i] = geopy.distance.geodesic(data, coor_supp).km
  sorted_indices = np.argsort(lst_coor)  # Sort the indices based on distances
  print(lst_coor)

  return (np.argmin(lst_coor), min(lst_coor), list(sorted_indices))
#Find nearest distance cluster from supplier 1
```

```python
nearest_cluster, nearest_distance,sorted_indices = findNearestCluster(np.array([-6.2088,106.8456]))
print(sorted_indices)

# cluster_locations = labeled_farmer_data[labeled_farmer_data['Cluster'] == nearest_cluster]
# cluster_locations.head()
```

```python
labeled_farmer_data['Rank'] = labeled_farmer_data['Cluster'].apply(lambda cluster: sorted_indices.index(cluster) + 1)
```
